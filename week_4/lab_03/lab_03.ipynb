```python
#create the movie class
#pass in dictionary object
class Movie():
    def __init__(self,movie_entry):
        self.title = movie_entry['Title']
        self.year = int(movie_entry['Year'])
        self.director = movie_entry['Director']
        self.actors = [x.strip() for x in movie_entry['Actors'].split(",")]
        #str(movie_entry['Actors'])
        #[x.strip() for x in movie_entry['Actors'].split(",")]
        self.rating = movie_entry['Rated']
        self.genre = [x.strip() for x in movie_entry['Genre'].split(",")]
        self.imdbrating = movie_entry['imdbRating']
    
    #str operator prints title and year
    def __str__(self):
        return str(self.title) + " - " + str(self.year) + " - " + str(self.director)
    
    #Method to check if actor is in film
    def actor_in(self, actor_name):
        #print(type(self.actors))
        if actor_name in set(self.actors):
            return True
        else:
            return False
    
    #Method to check similarity of Movies
    def similarity(self, other_movie):
        #print 'Checking similarity of %s and %s' % (self.title, other_movie.title)
        #instatitate similarity score
        sim_score = 0
        if self.rating == other_movie.rating:
            sim_score += 1
        if self.director == other_movie.director:
            sim_score += 1
        if self.year == other_movie.year:
            sim_score += 5
        if abs(self.year - other_movie.year) == 1:
            sim_score += 4
        if abs(self.year - other_movie.year) == 2:
            sim_score += 3
        if abs(self.year - other_movie.year) == 3:
            sim_score += 2
        if abs(self.year - other_movie.year) == 4:
            sim_score += 1
        if set(self.genre) & set(other_movie.genre):
            sim_score += int(len(set(self.genre) & set(other_movie.genre)))
        #remove the same film by setting score = 0
        if self.title == other_movie.title:
            sim_score = 0
        return sim_score
```

```python
#create movie recommender
class MovieRecommender():
    #initialise with list of movies to recommend from
    def __init__(self, movie_list):
        self.movie_list = movie_list
        
    #define movie recommender from above code snippets
    def recommend(self,movie):
        #create list of similarity scores
        similarity_scores = [movie.similarity(movie_l) for movie_l in self.movie_list]
        #find max score in list
        max_score = max(similarity_scores)
        #get list of movies with same similarity score
        boolean_list = [x == max_score for x in similarity_scores]
        filtered_list = [i for indx,i in enumerate(self.movie_list) if boolean_list[indx] == True]
        #pick a random movie from this list to return
        return (str(random.choice(filtered_list)),max_score)
```
